### Phase 1: Make a Tidy DataFrame

```python
# Function clean up filings into a tidy dataframe
def tidy_court_filings(df):
    """Make a tidy DataFrame of court filings
    
    Input a Pandas DataFrame from a CSV downloaded at
    https://opendata.maryland.gov/Housing/District-Court-of-Maryland-Eviction-Case-Data/mvqb-b4hf/about_data
    """
    # Drop unnecessary columns
    df = df.drop(columns=['Unnamed: 0'])
    
    # Convert column names to all lowercase with underscores instead of spaces
    # (This is a person style, but I find it easier to operate with column names that are simple and don't have spaces.)
    # One way to do this is with a for loop:
    new_column_names = []
    for col in df.columns:
        col = lower_underscore(col)
        new_column_names.append(col)
    df.columns = new_column_names
      
    # Make sure dates are stored as datetimes instead of strings
    df['event_date'] = pd.to_datetime(df['event_date'])
    df['evicted_date'] = pd.to_datetime(df['evicted_date'])

    # Make sure zip codes and years are integers
    # (Can you think of a situation where we wouldn't want to store them as integers?)
    df['tenant_zip_code'] = df['tenant_zip_code'].astype('Int64')
    df['tenant_zip_code'] = df['tenant_zip_code'].astype('Int64')

    # Make sure event types are consistently classified and use simple class names
    # (You can use the `map` method with a dictionary as the argument to reclassify
    # every value in a column that equals a key in the dictionary with value for that key.)
    reclassifier = {
        'Petition - For Warrant of Restitution Filed': 'petition',
        'petition - For Warrant of Restitution Filed': 'petition',
        'Warrant of Restitution - Return of Service - Cancelled': 'warrant_cancelled',
        'Warrant of Restitution - Return of Service - Evicted': 'warrant_evicted',
        'Warrant of Restitution - Return of Service - Expired': 'warrant_expired',
    }
    df['event_type'] = df['event_type'].map(reclassifier)

    # Make sure case types are consistently classified and use simple class names
    reclassifier = {}
    for case_type in df['case_type']:
        reclassifier[case_type] = lower_underscore(case_type)
    df['case_type'] = df['case_type'].map(reclassifier)

    # Make sure city names are consistently title case
    df['tenant_city'] = df['tenant_city'].str.title()

    # Recalculate event year and eviction year based on dates, just to be confident they're accurate
    # and make sure they're stored as integers
    df['event_year'] = df['event_date'].dt.year.astype('Int64')
    df['eviction_year'] = df['evicted_date'].dt.year.astype('Int64')

    # Change 'evicted_date' to 'eviction_date' for grammatical consistency with 'eviction_year'
    # (Okay, now I'm getting picky, but you can see how you can get into the details with making
    # a dataframe nice and tidy)
    df = df.rename(columns={'evicted_date': 'eviction_date'})

    # Put columns in a more intuitive order
    # (Again, highly optional, but can be useful for tidying)
    df = df[[
        'case_number',
        'case_type',
        'event_date',
        'event_year',
        'event_type',
        'event_comment',
        'eviction_date',
        'eviction_year',
        'county',
        # 'location', # I believe 'location' refers to the district court location, which isn't important to us, so let's leave it out
        'tenant_city',
        'tenant_state',
        'tenant_zip_code',
    ]]    

    return df

df_filings_tidy = tidy_court_filings(df_filings_raw)

df_filings_tidy.head(2)
```

### Phase 2: Analyze Tidy DataFrame to Answer Questions

#### 1. How many unique cases get filed each month?

```python
def unique_cases_per_month(df, distinguish_year=True):
    """Calculate unique cases filed in each calendar month

    df: Pandas DataFrame pre-processed by the `tidy_court_filings` function

    distinguish_year: True or False (default: True)
        If True, counts are broken out by year and month
        If False, counts from all years are combined in like calendar months
    """
    # Make sure that earliest events are first
    df = df.sort_values('event_date')
    # Keep only the first row for each case number
    unique_cases = df.drop_duplicates('case_number', keep='first')

    # Count cases in each month and year
    # adapted from https://stackoverflow.com/questions/38792122/how-to-group-and-count-rows-by-month-and-year-using-pandas
    if distinguish_year:
        case_counts = unique_cases.groupby([
            unique_cases['event_date'].dt.year.rename('year'), 
            unique_cases['event_date'].dt.month.rename('month')
        ])['case_number'].count()
    else:
        case_counts = unique_cases.groupby([ 
            unique_cases['event_date'].dt.month.rename('month')
        ])['case_number'].count()

    # Structure output in a dataframe with months and years as columns
    case_counts = pd.DataFrame(case_counts).reset_index()
    
    return case_counts
 
```

#### 2. What percent of cases end in eviction?

```python
#change: for some reason actual work didn't get uploaded so I'm retrying now 03/08/2025
#pseudocode: goal is to  define function to calculate warrant_evicted over count of unique id
#1: drop duplicate ids
#2: count cases by eviction status
#3: sum all case counts
#4: divide case counts by total count

def eviction_cases_prop(df):
    #drop duplicate cases
    unique_cases = df.drop_duplicates('case_number', keep='first')
    
    #group ids by eviction status & sum
    eviction_groups = unique_cases.groupby('event_type')['case_number'].count()
    
    #sum case counts
    case_sum = len(unique_cases)
    
    #divide eviction_groups by case counts
    eviction_props = eviction_groups / case_sum
    
    #add eviction_props as column
    eviction_props_df = pd.DataFrame(eviction_props).reset_index()
    
    return eviction_props_df

eviction_cases_prop(df_filings_tidy)
```

#### 3. Do cases of different types (e.g., tennant holding over, breach of lease, failure to pay rent) culminate in eviction at different rates?

```python
#pseudocode: goal is to  define function to calculate eviction levels grouped by different types
#1: drop duplicate ids
#2: count cases by eviction status
#3: sum all case counts and eviction counts
#4: create interior function for only evictions
#5: group evictions by case type
#6: divide eviction cases by total count

def eviction_types_prop(df):
    #drop duplicate cases
    unique_cases = df.drop_duplicates('case_number', keep='first')
    
    #group ids by eviction status & sum
    eviction_groups = unique_cases.groupby('event_type')['case_number'].count()
    
    #calculate total eviction count and case count
    eviction_count = len(unique_cases[unique_cases['event_type'] == 'warrant_evicted'])
    case_sum = len(unique_cases)

    #create function for evictions, grouping by eviction type
    if 'warrant_evicted' in unique_cases['event_type'].values:
        eviction_types = unique_cases[unique_cases['event_type'] == 'warrant_evicted']
        eviction_types = eviction_types.groupby('case_type')['case_number'].count()
        
    
    #divide eviction_groups by eviction counts and case counts
    eviction_type_props = eviction_types / eviction_count
    eviction_props = eviction_types / case_sum

    
    #add eviction_type_props as column
    eviction_type_props_df = pd.DataFrame(eviction_type_props).reset_index()
    
    #add eviction_props as column
    eviction_props_df = pd.DataFrame(eviction_props).reset_index()   
    
    return eviction_type_props_df, eviction_props_df

eviction_types_prop(df_filings_tidy)
```

#### 4. How long, on average, does it take for cases to move from a petition being filed to warrent of restitution?

```python
#pseudocode: goal to create function that determines length of cases being moved from petition to warrant

#1: drop duplicate ids
#2: ensure event date and eviction dates are datetime objects that can be calculated
#3: create column in dataframe that is calculation of event date - eviction date, in terms of days
#4: find average of new column

def eviction_petition_time(df):
    #drop duplicate cases
    unique_cases = df.drop_duplicates('case_number', keep='first')

    #ensure 'event_date' and 'eviction_date' are datetime objects
    unique_cases['event_date'] = pd.to_datetime(unique_cases['event_date'])
    unique_cases['eviction_date'] = pd.to_datetime(unique_cases['eviction_date'])
    
    #create column calculating time difference
    unique_cases['time_difference'] = (unique_cases['event_date'] - unique_cases['eviction_date']).dt.days

    #find average of time difference
    avg_time = unique_cases['time_difference'].mean()

    return avg_time


print(f'The average time for a case to move from a petition being filed to a warrent of restitution is {avg_time} days')
    
```

#### 5. What is the eviction rate per person, per year, at the county level? Which county has a higher rate?

```python
#pseduocode: goal is to define eviction rate per person, per year, at county level, and define who has a higher eviction rate
#1: import and join moco & pg census data
#2: remove all rows after first and columns that aren't population from census data, joining into one small df
#3: extract population data
#4: group eviction data by county and year, creating dictionaries for the county-year combos
#5: use loops to fill in dictionaries
#6: divide dicitionary counts by county population
#7: print results and compare

#import census data
mg_df = pd.read_csv('acs2023_5yr_B01003_mg.csv')
pg_df = pd.read_csv('acs2023_5yr_B01003_pg.csv')

#exclude non major rows and nonuseful columns
df_pop = pd.concat([mg_df, pg_df], axis=0)
df_pop = df_pop[df_pop['name'].isin(['Montgomery County, MD', 'Prince George\'s County, MD'])]
df_pop = df_pop.drop(columns=['geoid', 'B01003001, Error'])

#extract population data
moco_pop = df_pop.iloc[0]['B01003001']
pg_pop = df_pop.iloc[1]['B01003001']

#create eviction function
def eviction_rate(df, distinguish_year=True):
    
    #drop duplicates
    unique_cases = df.drop_duplicates('case_number', keep='first')

    # filter for only evictions, through 'warrant_evicted'
    eviction_cases = unique_cases[unique_cases['event_type'] == 'warrant_evicted']

    #ensure 'event_date' is a datetime object
    eviction_cases['event_date'] = pd.to_datetime(eviction_cases['event_date'])

    #filter cases by county
    moco_cases = eviction_cases[eviction_cases['county'] == 'Montgomery']
    pg_cases = eviction_cases[eviction_cases['county'] == "Prince George's"]

    #creates empty dictionaries forthe 4 groups
    case_counts_by_group = {
        'moco_2023': None,
        'moco_2024': None,
        'pg_2023': None,
        'pg_2024': None
    }

    years = [2023, 2024]
    
    # Loop over counties and years, calculating case counts for each
    for county_cases, county_name in [(moco_cases, 'moco'), (pg_cases, 'pg')]:
        for year in years:
            #filter by year
            yearly_cases = county_cases[county_cases['event_date'].dt.year == year]
            #count cases for each month of the year, if distinguishing by year
            if distinguish_year:
                case_counts = yearly_cases.groupby([
                    yearly_cases['event_date'].dt.year.rename('year')
                ])['case_number'].count()
            else:
                case_counts = yearly_cases.groupby('event_date')['case_number'].count()

            #store results in dictionaries
            case_counts_by_group[f'{county_name}_{year}'] = case_counts
    
    return case_counts_by_group
    

#yearly rates
moco_2023_rate = moco_2023 / moco_pop
moco_2024_rate = moco_2024 / moco_pop
pg_2023_rate = pg_2023 / pg_pop
pg_2024_rate = pg_2024 / pg_pop 

#statements
print(f'In 2023, Montgomery County saw {moco_2023.values[0]} evictions, which, given a population of {moco_pop}, results in a per person eviction rate of {moco_2023_rate.values[0]}')
print(f'In 2024, Montgomery County saw {moco_2024.values[0]} evictions, which, given a population of {moco_pop}, results in a per person eviction rate of {moco_2024_rate.values[0]}')
print(f'In 2023, Prince George/s County saw {pg_2023.values[0]} evictions, which, given a population of {pg_pop}, results in a per person eviction rate of {pg_2023_rate.values[0]}')
print(f'In 2024, Prince George/s County saw {pg_2024.values[0]} evictions, which, given a population of {pg_pop}, results in a per person eviction rate of {pg_2024_rate.values[0]}')


```
